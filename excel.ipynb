1.  Explain the difference between Absolute, Relative, and Mixed Cell
Referencing in Excel with examples.
- In Excel, cell referencing determines how formulas behave when copied or moved. Here’s the breakdown:

🔹 Absolute Referencing
Format: $A$1

Both column A and row 1 are fixed.

Copying the formula anywhere will always point to cell A1.

Example: =$A$1+5 → Always adds 5 to the value in A1.
Relative Referencing
Format: A1

Both column and row adjust relative to the new position.

Copying the formula one cell down changes A1 → A2.

Example: =A1+5 → If copied to the next row, it becomes =A2+5.

🔹 Mixed Referencing
Format: $A1 or A$1

One part is fixed, the other adjusts.

$A1 → Column A fixed, row changes when copied.

A$1 → Row 1 fixed, column changes when copied.
Example:

=$A1+5 → Always uses column A, but row shifts.

=A$1+5 → Always uses row 1, but column shifts.

2. What is a Macro in Excel? How does it help in automation?
- A Macro in Excel is essentially a recorded sequence of actions or instructions that can be replayed automatically. It’s written in VBA (Visual Basic for Applications) or created using the Macro Recorder.
What it is
A macro captures your steps (like formatting, calculations, or data manipulation).

Once saved, you can run it anytime to repeat those steps instantly.

🔹 How it helps in automation
Saves time: Repetitive tasks (e.g., formatting reports, applying formulas) can be done in one click.

Reduces errors: Manual repetition often leads to mistakes; macros ensure consistency.

Boosts productivity: Complex processes (like generating dashboards or cleaning data) can be automated.
🔹 Example
Suppose you prepare a monthly sales report:

Apply bold headers.

Format currency.

Insert a chart.

3.  What are Text Functions in Excel? Mention any five with examples.
- In Excel, Text Functions are built‑in formulas that help you manipulate and analyze text data. They’re especially useful for cleaning datasets, formatting values, or extracting information from strings.

Here are five common Text Functions with examples:

🔹 1. LEN()
Purpose: Counts the number of characters in a text string (including spaces).

Example: =LEN("Excel") → 5

🔹 2. LEFT()
Purpose: Extracts characters from the left side of a string.
Example: =LEFT("Analytics",4) → Anal

🔹 3. RIGHT()
Purpose: Extracts characters from the right side of a string.

Example: =RIGHT("Analytics",3) → ics

🔹 4. CONCAT() (or CONCATENATE() in older versions)
Purpose: Joins two or more text strings together.

Example: =CONCAT("Data","Science") → DataScience

🔹 5. UPPER() / LOWER()
Purpose: Converts text to uppercase or lowercase.

4.  What is the use of Scenario Manager in decision making?
- The Scenario Manager in Excel is part of the What‑If Analysis tools. It allows you to create and compare multiple sets of input values (scenarios) to see how they affect the outcome of formulas or models.

🔹 Use in Decision Making
Compare alternatives: You can build different business cases (e.g., optimistic, realistic, pessimistic) and instantly see how each affects profit, cost, or revenue.

Forecast outcomes: Helps managers test assumptions like changes in price, sales volume, or expenses.

Risk analysis: By storing multiple scenarios, you can evaluate the impact of best‑case vs worst‑case situations.
Strategic planning: Useful for budgeting, investment decisions, and project planning where multiple variables may change.

🔹 Example
Suppose you run a shop:

Scenario 1: Price = ₹100, Sales = 500 units → Revenue = ₹50,000

Scenario 2: Price = ₹120, Sales = 400 units → Revenue = ₹48,000

Scenario 3: Price = ₹90, Sales = 600 units → Revenue = ₹54,000

5. Define the purpose of VLOOKUP and HLOOKUP. How are they different
from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?
- VLOOKUP (Vertical Lookup)
Purpose: Searches for a value in the first column of a table and returns a value from another column in the same row.

Example: =VLOOKUP(101, A2:D10, 3, FALSE) → Finds ID 101 in column A and returns the value from the 3rd column of that row.

🔹 HLOOKUP (Horizontal Lookup)
Purpose: Searches for a value in the first row of a table and returns a value from another row in the same column.

Example: =HLOOKUP("Jan", A1:H5, 4, FALSE) → Finds “Jan” in row 1 and returns the value from the 4th row of that column.
XLOOKUP (Modern Replacement)
Purpose: More flexible, replaces both VLOOKUP and HLOOKUP.

Can search both vertically and horizontally.

Allows searching from any column/row, not just the first.

Handles missing values better (with default return options).

Example: =XLOOKUP(101, A2:A10, C2:C10, "Not Found") → Finds ID 101 in column A and returns corresponding value from column C.

🔹 INDEX-MATCH
Purpose: Combination of two functions:

INDEX() returns a value from a table based on row/column number.

MATCH() finds the position of a value in a row/column.

Together, they provide powerful, flexible lookups.

Example: =INDEX(C2:C10, MATCH(101, A2:A10, 0)) → Finds ID 101 in column A and returns corresponding value from column C.
